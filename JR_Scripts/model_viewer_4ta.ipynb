```python
def gen_data(num_classes, n):
    num_gaussians = 42
    if 'num_gaussians' in params:
        num_gaussians = params['num_gaussians']

    # Generate data
    data = time_toy_generator.gen_dataset_continuous(images_per_time_step=n,
                                      width=params['image_size'][0],
                                      num_gaussians=num_gaussians,
                                      point_density_factor=3)
    if num_classes == 8:
        data = np.asarray([data[2], data[3], data[4], data[5], data[6], data[7], data[8], data[9]])
    if num_classes == 5:
        data = np.asarray([data[1], data[3], data[5], data[7], data[9]])
    if num_classes == 4:
        data = np.asarray([data[0], data[3], data[6], data[9]])
    if num_classes == 2:
        data = np.asarray([data[5], data[9]])
    if num_classes == 1:
        data = np.asarray([data[9]])

    # Prep data
    data = data.swapaxes(0,1)
    data = data.reshape((data.shape[0] * data.shape[1], data.shape[2], data.shape[3]))
    data = data.astype(np.float32)
    data = utils.forward_map(data, params['cosmology']['k'])
    data = data / 1.02
    return data
def get_cmap(name, n_colors):
    cmap = cm.get_cmap(name, n_colors)
    return cmap(np.arange(n_colors))
cmap = get_cmap('winter', 10)
```

```python
data = gen_data(10, 100)
```

```python
# Initialize model
if params['model_idx'] == 0:
    model = WGanModel
if params['model_idx'] == 1:
    model = WNGanModel
if params['model_idx'] == 2:
    model = TemporalGanModelv3
```

```python
gan = GAN(params, model)
gan._sess = tf.Session()
leo_results = "/home/jonathan/Documents/Master_Thesis/LeoResults/"
folder = "models_M23/TWGAN_TS4_gp50_2018_3_21_15_33_checkpoints/"
specific_model = "TWGAN_TS4_gp50-232000"
gan._load(leo_results + folder + specific_model)
```

```python
def gen_images_and_add_to_list(lst, width = 128):
    z = gan._sample_latent(1)
    frames = (np.arange(10) + 3) / 12
    for idx in frames:
        y = np.asarray([[idx]])
        g_fake = gan._sess.run([gan._G_fake], feed_dict={gan._z:z, gan._model.y:y})[0]
        img = np.reshape(g_fake, (width, width))
        lst.append(img)
        
def wasserstein_distance(x_og, y, w):
    assert(x_og.shape == y.shape == w.shape)
    x = np.copy(x_og)
    loss = 0
    for idx in range(x.shape[0] - 1):
        d = y[idx] - x[idx]
        x[idx] = x[idx] + d
        x[idx+1] = x[idx+1] - d
        loss = loss + np.abs(d * (w[idx+1] - w[idx]))
    return loss / (w[-1] - w[0])
```

```python
img_series = []
width = params["image_size"][0]
gen_images_and_add_to_list(img_series, width = width)
fig, ax = plt.subplots(figsize=(128, 16))
ax.imshow(np.vstack([np.hstack(img_series), np.hstack(data[0:10])]), interpolation=None)
plt.tight_layout()
```

```python
images = []
for i in range(100):
    gen_images_and_add_to_list(images, width = width)
```

# Histogram of Data

```python
_, bins = np.histogram(np.vstack((data, images)), bins=20)
a = (bins[1:] + bins[:-1]) / 2
plt.figure()
for i in range(10):
    hist_r,_ = np.histogram(data[i:10:], density=True, bins=bins)
    plt.plot(a, hist_r, '-', c=cmap[i])
plt.yscale("log")
plt.figure()
for i in range(10):
    hist_f,_ = np.histogram(images[i:10:], density=True, bins=bins)
    plt.plot(a, hist_f, '-', c=cmap[i])
plt.yscale("log")
```

```python
for t in range(10):
    plt.figure()
    hist_f, bins = np.histogram(images[t:10:], density=True, bins=20)
    hist_r, _ = np.histogram(data[t:10:], density=True, bins=bins)
    x = (bins[1:] + bins[:-1]) / 2
    plt.plot(x, hist_f, '-', label="fake")
    plt.plot(x, hist_r, '-', label="real")
    print("t{} Log L2 Norm: {}".format(t, np.linalg.norm(hist_r-hist_f)))
    print("t{} Wasserstein: {}\n".format(t, wasserstein_distance(hist_f, hist_r, x)))
    plt.legend(loc=1)
    plt.title("t{}".format(t))
    plt.yscale("log")
```

# Temporal Comparison

```python
diffs_r = data[1:] - data[:-1]
data_f = np.asarray(images)
diffs_f = data_f[1:] - data_f[:-1]
```

```python
_, bins = np.histogram(np.vstack((diffs_r, diffs_f)), bins='doane')
a = (bins[1:] + bins[:-1]) / 2
plt.subplot(1,2,1)
plt.title("Real")
for i in range(10-1):
    hist_dr = np.histogram(diffs_r[i:10:], density=True, bins=bins)
    plt.plot(a, hist_dr[0], '-', c=cmap[i])
plt.subplot(1,2,2)
plt.title("Fake")
for i in range(10-1):
    hist_df = np.histogram(diffs_f[i:10:], density=True, bins=bins)
    plt.plot(a, hist_df[0], '-', c=cmap[i])
```

```python
for t in range(10-1):
    plt.figure()
    plt.title("t{}-t{}".format(t+1, t))
    hist_dr, _ = np.histogram(diffs_r[t:10:], density=True, bins=bins)
    hist_df, _ = np.histogram(diffs_f[t:10:], density=True, bins=bins)
    x = (bins[1:] + bins[:-1]) / 2
    print("t{}-t{} Wasserstein distance {}".format(t+1,t,wasserstein_distance(hist_df, hist_dr,x)))
    plt.plot(x, hist_dr, '-', label="real")
    plt.plot(x, hist_df, '-', label="fake")
    plt.legend(loc=2)
```
